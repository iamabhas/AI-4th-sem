```python
# VACCUM CLEANER PROGRAM :

# importing random library from python
import random

class Rooms:
    def __init__(self):
        
        # Determines if Room's are Clean or Dirty

        self.states ={
            'Location A': random.choice(['Clean', 'Dirty']),
            'Location B': random.choice(['Clean','Dirty'])
            }

        # Checks if Room A is Dirty or not and cleans it if it is 

    def refresh(self,room):
            if room =='Location A':  
                self.states = {
                "Location A" : random.choice(['Clean','Dirty'])
                }

        # Checks if Room B is Dirty or not and cleans it if it is
            
            elif room== "Location B":
                self.states = {
                'Location B': random.choice(['Clean','Dirty'])
                }
        
        # Does the same for both Room's

            else:
                self.states ={
                'Location A': random.choice(['Clean', 'Dirty']),
                'Location B': random.choice(['Clean','Dirty'])
                }
   
                
class TheVaccumCleaner:
    def __init__(self):
        self.location = random.choice (['Location A', 'Location B'])
        self.room = Rooms()
        self.cost = 0
        
    def startVaccum(self):
        self.Action('cleaning...!')
        self.room.states[self.location]= 'Clean'
        self.cost +=1
        
    def moveVaccum(self,direction):
        self.Action(direction)
        if(direction =='left'):
            self.location= 'Location A'
        else:
            self.location= 'Location B'
            
        # If the vaccum moves update the cost by 1

        self.cost+= 1
        
    def Action(self, action):

        # setting up if statements for three different actions which are cleaning , moving to leff/right and if there is no operation
        
        if action=="cleaning...!":
            if(self.location=="Location A"):
                print(f"Currently Cleaning Room A...")
            else:
                print(f"Currently Cleaning Room B...")
            
        if action=='left' or action=='right':
            print(f"Currently in {self.location} and moving {action}...")
        if action=="no operation":
            print(f"NO OPERATION DETECTED! : Both Rooms are clean so vaccum dosent need to move . The current cost is {self.cost}")


    # This function highlights whether the vaccum is active or not 

    def state(self):
        if self.room.states['Location A']=='Clean' and self.room.states['Location B']=='Clean':
            print("BOTH ROOMS ARE CLEAN : Vaccum cleaner is inactive !! ")
        else:
            print("STILL ONE DIRTY ROOM DETECTED : Vaccum cleaner is active !! ")
    

    # This function checks if the vaccum has no operations meaning both room are cleana and it has no where to move so action becomes of 'no operation'

    def noOperation(self):
            self.Action('no operation')


    def startOperation(self):

        if(self.location =='Location A'):
            if self.room.states['Location A']== 'Dirty':
                print("Room A is dirty.")
                self.state()
                self.startVaccum()
                print(f"The Cost after cleaning Room A is {self.cost}")
                self.moveVaccum('right')
                self.room.refresh("A")  
                print(f"The Cost after moving right(to Room B) is {self.cost}")
                print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
                
            else:
                print("Room A is clean")
                self.state() 
                if self.room.states['Location A']=='Clean' and self.room.states['Location B']=='Clean':
                    self.noOperation()
                else:
                    self.moveVaccum('right')
                    print(f"The Cost after moving right(to Room B) is {self.cost}")
                
                print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
                
        elif(self.location == 'Location B'):
                if self.room.states['Location B'] == 'Dirty':
                    print("Room B is dirty.")
                    self.state()
                    self.startVaccum()
                    print(f"The Cost after cleaning Room B is {self.cost}")
                    self.moveVaccum('left')
                    self.room.refresh("B")
                    
                    print(f"The Cost after moving left(to Room A) is {self.cost}")
                    print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
                else:
                    print("Room B is clean")
                    self.state() 
                    if self.room.states['Location A']=='Clean' and self.room.states['Location B']=='Clean':
                        self.noOperation()
                    else:
                        self.moveVaccum('left')
                        print(f"The Cost after moving left(to Room A) is {self.cost}")
                    
                    print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
                
        else: 
            print("Vaccum is out of bounds !")
            
            

# creating instance of class 
agent= TheVaccumCleaner()

#Asking user for iterations the vaccum should perform its task
iterations=input("Enter number of iterations: ")

print("VACCUM CLEANER PROGRAM :\n")

#Iterating according to the users input
for i in range(int(iterations)):
    #calling the method from the class
    agent.startOperation()


```
