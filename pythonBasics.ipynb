```python
# 3. If a five-digit number is input through the keyboard, write a program to reverse the number.
def Reverse(number):

    # Asks for number again if user enters a number other than a 5 digit one a does not stop until correct input is provided
    while len(str(number))!=5:
        number = int(input("Thats not a five digit number . Enter again ! "))

    reverse = 0    
    while(number > 0):    
        rem = number%10    
        reverse = (reverse *10) + rem    
        number = number//10
    
    return f'Reverse of given number is {reverse}'


number = int(input("Enter a five digit number "))


print(Reverse(number))
```

```python
# 4. If a four-digit number is input through the keyboard, write a program to obtain the sum of the first and last digit of this number.
def SumOfFirstAndLast(number):

    # Asks for number again if user enters a number other than a 4 digit one a does not stop until correct input is provided
    while len(str(number))!=4:
        number = int(input("Thats not a five digit number.Enter again ! "))

    sum = int(str(number)[0]) + int(str(number)[len(str(number))-1])
    
    return f'The sum of first and last digit of given number is {sum}'


number = int(input("Enter a four digit number "))


print(SumOfFirstAndLast(number))    
```

```python
# 5. Write a program to read an amount in rupee and break the amount into possible number of bank notes as bank notes are of Rs. 1000, 100, 10 and 1 only.
bankNotes = (1000, 100, 10, 1)
amount = int(input("Enter amount in rupee "))
possibleBankNotes = {}
for n in bankNotes:
    possibleBankNotes[n] = amount // n
    amount %= n


for key,value in possibleBankNotes.items():
	print(f'There are {value} , {key} rupee notes')
```

```python
# 6. Two numbers are input through the keyboard into two locations C and D. Write a program to interchange the contents of C and D.
def Interchange(C,D):
    print(f"Before Swapping : Content of location C = {C} , Content of location D = {D}")
    temporaryLocation = C
    C=D
    D=temporaryLocation
    print(f"After Swapping : Content of location C = {C} , Content of location D = {D}")


location_C=int(input("Enter contents of location C "))
location_D=int(input("Enter contents of location D "))

Interchange(location_C,location_D)
```

```python
""" 18. Given the coordinates (x, y) of a center of a circle and itâ€™s radius, 
write a program which will determine whether a point lies inside the circle, on the circle or outside the circle. """

x1=float(input('Enter center coordinate x1: '))
y1=float(input('Enter center coordinate y1: '))
radius=float(input('Enter the radius of the circle: '))
x2=float(input('Enter point coordinates x2: '))
y2=float(input('Enter point coordinates y2:'))
circle=(pow(x2-x1,2)+pow(y2-y1,2))
if(circle>radius):
  print("Point lies outside the circle")
elif(circle<radius):
  print ("Point lies inside the circle")
elif(circle==radius):
  print("Point lies on the boundary of circle")
else:
  print("Invalid")    
```

```python
""" According to the Gregorian calendar, it was Monday on the date 01/01/1900. 
If any year is input through the keyboard write a program to find out what is the day on 1st January of this year."""

year=float(input("Enter the year: "))
defaultYear=1900
year=(year-1)-defaultYear
leapYear=year/4 
yearLeft=year-leapYear
totalDays = (yearLeft*365)+(leapYear*366)+1
day = totalDays % 7
if(day==0):
  print("Monday")
elif(day==1):
  print("Tuesday")
elif(day==2):
  print("Wednesday")
elif(day==3):
  print("Thursday")
elif(day==4):
  print("Friday") 
elif(day==5):
  print("Saturday") 
elif(day==6):
  print("Sunday")
else:
  print("Invalid")   
   
```
